```python
import googlemaps
```

```python
def get_key(fpath="key.txt"):
    try:
        with open(fpath, 'r') as file:
            key = file.read().strip()
        return key
    except FileNotFoundError:
        return "File not found"
```

```python
gmaps = googlemaps.Client(key=get_key())
```

```python
import pandas as pd
```

```python
df = pd.read_csv("data/Police_Traffic_Crash_Reports.csv")
```

```python
df.shape
```

```python
import json

def geocode(address):
    return gmaps.geocode(address)

def make_address(entry):
    return entry + ", Virginia Beach, VA"

def format_response(response):
    return json.dumps(response[0])

def get_response(entry):
    response = geocode(make_address(entry))
    return format_response(response)
```

```python
class SavingDataframe:
    def __init__(self, cols, fpath="data/saved_df.csv"):
        self.df = pd.DataFrame(columns=cols)
        self.file_path = fpath
    
    def add(self, row_data):
        if len(row_data) == len(self.df.columns):
            self.df.loc[len(self.df)] = row_data
            self.df.to_csv(self.file_path, index=False)
        else:
            print("Invalid Data Size")
```

```python
data = SavingDataframe(["address", "response"])
```

```python
import time
for i, address in enumerate(df["Main_Street"]):
    response = get_response(address)
    data.add([address, response])

    print(f"Completed request {i}. Address : {address}")

    time.sleep(1.3)
    
    if i > 1000:
        break
```
