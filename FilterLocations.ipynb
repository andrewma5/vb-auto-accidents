```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data/FINAL_Police_Traffic_Crash_Reports.csv")
data.head()
```

```python
data.dropna(subset=['latitude', 'longitude'], inplace=True)
```

```python
# After our first checkpoint, we realized that there were several duplicate points mapping to the same location, as can be seen below
```

```python
location_counts = data[["latitude", "longitude"]].value_counts()
location_counts
```

```python
# We suspect that this is because several rows do not have a block number, the below cell confirms our suspicions
```

```python
index = 0
repeats = (data["latitude"] == location_counts.index[index][0]) & (data["longitude"] == location_counts.index[index][1])
data[repeats].head()
```

```python
vb_blvd = data[data["Main_Street"].str.contains("virginia beach", case=False)]
vb_blvd.head()
```

```python
vb_blvd_with_number = data[data["Main_Street"].str.contains(r"\d* virginia beach", case=False)]
vb_blvd_with_number.head()
```

```python
len(vb_blvd), len(vb_blvd_with_number)
```

```python
# 786 virginia beach blvd columns still have the block number
```

```python
# let's try filtering for all streets that do not have a road number
```

```python
filtered = data[data["Main_Street"].str.contains(r"\d+", case=False)]
filtered
```

```python
filtered_counts = filtered[["latitude", "longitude"]].value_counts()
filtered_counts
```

```python
# There still seem to be some duplicates
```

```python
index = 0
general_vb_repeats = (filtered["latitude"] == filtered_counts.index[index][0]) & (filtered["longitude"] == filtered_counts.index[index][1])
filtered[general_vb_repeats].head()
```

```python
# This location seems to be the location for when google maps does not recognize the location and goes to VB in general, we can also drop these
```

```python
filtered = filtered[general_vb_repeats == 0]
len(filtered)
```

```python
filtered_counts = filtered[["latitude", "longitude"]].value_counts()
repeats = (filtered["latitude"] == filtered_counts.index[0][0]) & (filtered["longitude"] == filtered_counts.index[0][1])
filtered[repeats].head()
```

```python
# The one above seems like a genuinely dangerous intersection
```

```python
filtered_counts = filtered[["latitude", "longitude"]].value_counts()
repeats = (filtered["latitude"] == filtered_counts.index[1][0]) & (filtered["longitude"] == filtered_counts.index[1][1])
filtered[repeats].head()
```

```python
# Same here! We will check one more to confirm we are done filtering
```

```python
filtered_counts = filtered[["latitude", "longitude"]].value_counts()
repeats = (filtered["latitude"] == filtered_counts.index[2][0]) & (filtered["longitude"] == filtered_counts.index[2][1])
filtered[repeats].head()
```

```python
# Looks good!
```

```python
filtered.to_csv("data/FILTERED_Police_Traffic_Crash_Reports.csv", index=False)
```
